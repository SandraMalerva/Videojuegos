##  Prepara los datos 

###  Reemplaza los nombres de las columnas

```python
games.columns= games.columns.str.lower()
```

```python
games
```

### Elige la manera de tratar los valores ausentes:


```python
print(games.isna().sum())
```

```python
games['user_score'] = games['user_score'].fillna(0)
```

```python
games['critic_score'] = games['critic_score'].fillna(0)
```

### Convierte los datos en los tipos necesarios.

```python
games['critic_score'] = games['critic_score'].astype(int)
```

```python
games['user_score'] = pd.to_numeric(games['user_score'], errors='coerce')
```

```python
games['user_score'] = games['user_score'].astype(float)
```

```python
games[games['year_of_release'].isnull()]
```

```python
 games['year_of_release'].mean()
```

```python
games['year_of_release'] = games['year_of_release'].fillna(2006)
```

```python
games['year_of_release'] = games['year_of_release'].astype(int)
```

```python
games
```

Se realizaron algunos cambios en 3 de las columnas, para dos de ellas fue necesario un convertir a un tipo de datos int, mientras que para la tercera fue un tipo de dato float.  
 
En la columna 'critic_score' se realizo la conversión a tipo **int** ya que contiene solo numeros y el rango es de 0 a 100 sin ningun valor decimal.  



Para la comumna 'year_of_release' que contenia 269 valores ausentes y para poder convertirla a tipo int64 fue necesario sustituir los valores ausentes por un número, en este caso fue 2006 ya que se tomo del valor promedio del año de lanzamiento, posteriormente se realizo la conversión a tipo **int** ya que contiene solo números.


En la columna 'user_score' se utilizo **errors='coerce'** para convertir los valores no numéricos en NaN, puesto que contenia valores 'tbd lo cual no nos permitia cambiar el tipo de dato,  posteriormente se realiza la conversión a tipo float ya que tiene un rango 0 a 10 y en algunos puntajes se encuentran valores decimales como 8.3.



```python
games['total_sales'] = games['na_sales'] + games['jp_sales'] + games['eu_sales'] + games['other_sales']
```

##  Analiza los datos

### Mira cuántos juegos fueron lanzados en diferentes años.

```python
lanzamiento_por_año = games['year_of_release'].value_counts().sort_index()
```

```python
lanzamiento_por_año
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=lanzamiento_por_año, marker='o')
plt.title('Lanzamiento de juegos por año')
plt.xlabel('Año de lanzamiento')
plt.ylabel('Cantidad de juegos')
plt.xticks(range(1980, 2017))
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

### Observa cómo varían las ventas de una plataforma a otra

```python
platform_sales = games.groupby('platform')['total_sales'].sum().sort_values(ascending=False)
```

```python
top_platform = platform_sales.head(5)
top_platform
```

```python
for platform in top_platform.index:
    ventas_anuales = games[games['platform']==platform].groupby('year_of_release')['total_sales'].sum()
    plt.plot(ventas_anuales.index, ventas_anuales.values, label=platform)
    
plt.title('Distrubución de ventas por año de las plataformas actuales más vendidas')
plt.xlabel('Año de lanzamiento')
plt.ylabel('Ventas totales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plataforma_ventas_por_año = games.groupby('platform').agg({"year_of_release":[min,max]}).reset_index()

plataforma_ventas_por_año.columns=["platform","min_year_of_release","max_year_of_release"]
```

```python
plataforma_ventas_por_año['duracion'] = plataforma_ventas_por_año['max_year_of_release'] - plataforma_ventas_por_año['min_year_of_release']
```

```python
plataforma_ventas_por_año['tipo']=plataforma_ventas_por_año['min_year_of_release'].map(lambda x: "nueva" if x>=2000 else "antigua")
```

```python
populares=list(plataforma_ventas_por_año[plataforma_ventas_por_año["max_year_of_release"]<2000]["platform"].unique())
```

```python
for platform in populares:
    ventas_anuales = games[games['platform']==platform].groupby('year_of_release')['total_sales'].sum()
    plt.plot(ventas_anuales.index, ventas_anuales.values, label=platform)
    
plt.title('Plataformas que solían ser populares pero que ahora no ')
plt.xlabel('Año de lanzamiento')
plt.ylabel('Ventas totales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
promedio_en_desaparecer = plataforma_ventas_por_año.groupby('tipo')['duracion'].mean()
promedio_en_desaparecer
```

Se contruyo un dataframe con la plataforma, el primer año de venta y el ultmo año de venta registrada, posteriormente se añadio una columna con la duración, una mas que contiene informacion sobre si es nueva o antigua tomando el año 2000 como referencia, menor que 2000 se tomo como antigua mientas que si es mayor se definio como nueva. 

Apartir de 200 hasta 2016 deltop 5 de los nuenvos juegos que se lanzaron, alcanzaron ventas entre los 140 y 250 millones de dorales.
Mientras que en las plataformas que solían ser populares alcanzaron valores en ventas mas altas de 40  a 60 millones de dplares, este periodo va de 1982 hasta 1999, dentro de este periodo 6 de los 7 mas populares se encuentan entre 1990 y 1999 algunos con 4 años, otros mas con dos y el de mayor duración es de 9 años, la mayoria terminando con ventas por debajo de los 10 millones. 
POor lo tanto los datos de cada período si son significativos

 Generalmente las plataformas antiguas tardan en desaparecer 9.5  años


Dada la informacion anterior creemos relevante solo considerar las plataformas que seguian teniendo ventas hasta el año 2000, todas las empresas que dejaron de tener ventas antes o en ese año no seran consideradas.

### ¿Qué plataformas son líderes en ventas?

```python
plataformas_actuales = games[~games["platform"].isin(populares)]
plataformas_actuales
```

```python
plataforma_ventas_actuales_por_año = plataformas_actuales.groupby('platform').agg({"year_of_release":[min,max], 'total_sales' : 'sum'}).reset_index()
```

```python
plataforma_ventas_actuales_por_año.columns=["platform","min_year_of_release","max_year_of_release", "total_sales_sum"]
plataforma_ventas_actuales_por_año = plataforma_ventas_actuales_por_año.sort_values(by=['total_sales_sum'], ascending=False)
```

```python
plataforma_ventas_actuales_por_año['duracion'] = plataforma_ventas_actuales_por_año['max_year_of_release'] - plataforma_ventas_actuales_por_año['min_year_of_release']

plataforma_ventas_actuales_por_año
```

```python
plataformas_mas_rentables =plataforma_ventas_actuales_por_año[plataforma_ventas_actuales_por_año['max_year_of_release'] == 2016]
```

```python
plataformas_mas_rentables
```

```python
plataformas_actuales=list(plataformas_mas_rentables["platform"].unique())
plataformas_actuales
```

```python
for platform in plataformas_actuales:
    ventas_anuales = games[games['platform']==platform].groupby('year_of_release')['total_sales'].sum()
    plt.plot(ventas_anuales.index, ventas_anuales.values, label=platform)
    
plt.title('Plataformas son líderes en ventas')
plt.xlabel('Año de lanzamiento')
plt.ylabel('Ventas totales')
plt.legend()
plt.grid(True)
plt.show()
```

¿Qué plataformas son líderes en ventas?
Wii,PS3, X360 son las plataformas lideres en ventas, que estan arriba de los 150 millones de dolares, aunque tuvieron buenas ventas y porteriormente bajaron se mantienen, mientas que PS4, XOne, WiiU, tuvieron pocas vetas y solo dos años de vida. 
Por otro lado la plataforma PC cuenta con  pocas ventas pero se ha mantenido durante 31 años. 
las mas rentables X360 y PS3

### Gráfico de dispersión

```python
ds = games[games['platform'] == 'DS'].sort_values('year_of_release')
ds
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=ds, x='critic_score', y='total_sales')
plt.title('Ventas vs Puntuación de Críticos (DS)')
plt.xlabel('Puntuación de Críticos')
plt.ylabel('Ventas Totales (millones)')
plt.grid(True)
plt.show()


plt.figure(figsize=(10, 5))
sns.scatterplot(data=ds, x='user_score', y='total_sales')
plt.title('Ventas vs Puntuación de Usuarios (DS)')
plt.xlabel('Puntuación de Usuarios')
plt.ylabel('Ventas Totales (millones)')
plt.grid(True)
plt.show()

critic_corr = ds['critic_score'].corr(ds['total_sales'])
user_corr = ds['user_score'].corr(ds['total_sales'])

print(f'correlacion entre críticos y ventas: {critic_corr:.2f}')
print(f'correlacion entre usuarios y ventas: {user_corr:.2f}')
```

Ambas correlaciones son similares y muy bajas, esto significa que ni las críticas profesionales ni las valoraciones de usuarios tienen una relación fuerte con las ventas.

## Prueba las siguientes hipótesis

```python
xbox_one=games[games['platform']=='XOne'].dropna()['user_score']
pc=games[games['platform']=='PC'].dropna()['user_score']

alpha = 0.05

results = st.ttest_ind(xbox_one, pc, equal_var=False)
print('valor p:', results.pvalue)

if (results.pvalue < alpha ):
    print("Rechazamos la hipótesis nula")
else:
    print("No podemos rechazar la hipótesis nula")
```

- Hipotesis nula: Las medias de las puntuaciones de usuario para juegos de Xbox One y PC.c son iguales.
- Hipotesis alternativa: Las medias de las puntuaciones de usuario para juegos de Xbox One y PC si tienen una diferencia significativa.
- Dado que el valor p obtenido fue muy pequeño, se rechaza la hipótesis nula, concluyendo que las medias son distintas.

```python
accion=games[games['genre']== 'Action'].dropna()['user_score']
deportes=games[games['genre']=='Sports'].dropna()['user_score']
```

```python

alpha = 0.05

results = st.ttest_ind(accion, deportes, equal_var=False)
print('valor p:', results.pvalue)

if (results.pvalue < alpha ):
    print("Rechazamos la hipótesis nula")
else:
    print("No podemos rechazar la hipótesis nula")
```

- Hipotesis nula: Las medias de de las puntuaciones de usuario para juegos del género Acción y Deportes son iguales.
- Hipotesis alternativa: Las medias  de las puntuaciones de usuario para juegos de Acción y Deportes tienen una diferencia significativa.
- Si el valor p (pvalue) es menor que 0.05, rechazas la hipótesis nula, dado que en este caso, el valor p resultó ser 0.127, que es mayor a 0.05, por lo que no se rechazo la hipótesis nula, lo cual indica que no hay una diferencia significativa entre las puntuaciones de usuario de los juegos de Acción y Deportes.
